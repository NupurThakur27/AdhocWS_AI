```
#importing modules
from bs4 import BeautifulSoup
import requests #replacement for urllib
import time
# corona live data set
url='https://www.worldometers.info/coronavirus/'
#  connecting with  https protocol
htmldata=requests.get(url)
# print(htmldata.text)
print(htmldata.content)
```

```
purehtmldata=htmldata.content
souped_data=BeautifulSoup(purehtmldata,'html5lib') #souping
table_data=souped_data.find_all('table',class_="table table-bordered table-hover main_table_countries")
# print(table_data)
# finding all table rows
for trr in table_data[0].find_all('tr'):
  for headers in trr.find_all('th'):
    print(headers.text)
```

```
# Data for each row with all Country
for trr in table_data[0].find_all('tr'):
  for rows in trr.find_all('td'):
    print(rows.text)
```
