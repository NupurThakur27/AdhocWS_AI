```
# loading matplotlib library
#Learning Data Visualization
import matplotlib.pyplot as plt

x=[2,9,3,12] 
y=[7,15,5,9]
# Coordniates are (2,7) and (9,15)
plt.xlabel("Time",color="blue") # x-axis
plt.ylabel("Distance",color="blue") #y-axis
plt.plot(x,y,color="green",label="RACE TRACK") # This will draw a line
plt.scatter(x,y,s=200,label="STOP POINTS",marker="*") #This will draw (default marker is a filled circle) points, point size can be changed by s parameter.
plt.bar(x,y,label="Distance covered by cars") #To draw bar charts
plt.grid(color="blue") # this is will draw the grid
plt.legend() #It will show the labels of scatter and plot
plt.show() # it will draw graph
```

```
#This will be another example of Cricket Data Visualization
players=["Dhoni","Virat Kohli","Sachin","KL Rahul"]
scores=[120,175,100,140]


#plotting the above data
plt.xlabel("Players")
plt.ylabel("Scores")
# plt.bar(players,scores,label="2020 Score of Indian Cricket Team")
plt.pie(scores,labels=players, explode=(0.25,0,0,0),shadow=True,autopct='%1.2f%%') #to draw pie chart
#plt.grid(color="green")
plt.legend()
plt.show()
```
